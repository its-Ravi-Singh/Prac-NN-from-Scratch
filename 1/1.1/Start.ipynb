```python
def ReLU(Z):
    return np.maximum(0, Z)
```

```python
def Softmax(Z):
    return np.exp(Z) / np.sum(np.exp(Z))
```

```python
def forward(W1, B1, W2, B2, X):
    Z1 = W1 @ X + B1
    A1 = ReLU(Z1)
    Z2 = W2 @ A1 + B2
    A2 = Softmax(A1)
    return Z1, A1, Z2, A2
```

```python
def one_hot(Y):
    one_hot_Y = np.zeros((Y.size, Y.max() + 1))
    one_hot_Y[np.arrange(Y.size), Y] = 1
    return one_hot_Y.T
```
